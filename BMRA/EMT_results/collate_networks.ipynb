```python
import pandas as pd
import numpy as np
from numpy import NaN
import matplotlib.pyplot as plt
import os
import re
```

```python
files_list = os.listdir()
relist = re.compile(r'(^[A-Z0-9]+)_rm_log_200_5K_([A-Za-z]+).csv')

for filename in files_list:
        srch = relist.search(filename)
        if srch:
            matrix_r = pd.read_csv(filename)
            matrix_r.index = matrix_r.columns
            for module_from in matrix_r.columns:
                for module_to in matrix_r.index:
                    if module_from == module_to:
                        matrix_r.at[module_to,module_from] = -1
            cell_line = srch.group(1)
            ligand = srch.group(2)
            matrix_rs = pd.read_csv(cell_line + '_rs_log_200_5K_' + ligand + '.csv')
            matrix_rs.index = matrix_rs.columns
            matrix_combine = matrix_r.round(decimals=2).astype(str) + ' ± ' + matrix_rs.round(decimals=2).astype(str)
            for module_from in matrix_combine.columns:
                for module_to in matrix_combine.index:
                    if module_from == module_to:
                        matrix_combine.at[module_to,module_from] = '-1'
                    if matrix_r.at[module_to,module_from] == 0:
                        matrix_combine.at[module_to,module_from] = '0'
            matrix_r.to_excel(cell_line + '_rm_log_200_5K_' + ligand + '_ed.xlsx')
            matrix_rs.to_excel(cell_line + '_rs_log_200_5K_' + ligand + '_ed.xlsx')
            matrix_combine.to_excel(cell_line + '_rm_log_200_5K_' + ligand + '_comb.xlsx')
```

```python
filename = 'A549_rm_log_200_5K_TGFb_ed.xlsx'
matrix_r = pd.read_excel(filename,index_col=0)

files_list = os.listdir()
relist = re.compile(r'(^[A-Z0-9]+)_rm_log_200_5K_([A-Za-z]+)_ed.xlsx')

cn_columns = []
for filename in files_list:
        srch = relist.search(filename)
        if srch:
            cell_line = srch.group(1)
            ligand = srch.group(2)
            cn_columns.append(cell_line + ' + ' + ligand)
cn_columns.sort()

cn_rows = []
for module_from in matrix_r.columns:
    if module_from != 'DPD':
        for module_to in matrix_r.index:
            if module_from != module_to:
                cn_rows.append(module_from + ' -> ' + module_to)

connections_rm = pd.DataFrame(np.zeros((len(cn_rows),4*3)),index=cn_rows,columns=cn_columns)
connections_rs = pd.DataFrame(np.zeros((len(cn_rows),4*3)),index=cn_rows,columns=cn_columns)
connections_rm
```

```python
for filename in files_list:
        srch = relist.search(filename)
        if srch:
            cell_line = srch.group(1)
            ligand = srch.group(2)
            matrix_rm = pd.read_excel(filename,index_col=0)
            matrix_rs = pd.read_excel(filename.replace('_rm_','_rs_'),index_col=0)
            # filling textual matrix
            matrix_text = []
            matrix_math = []
            for module_to in matrix_rm.index:
                matrix_row=[]
                matrix_math_row=[]
                for module_from in matrix_rm.columns:
                    if module_from == module_to:
                        matrix_row.append('-1')
                        matrix_math_row.append(-1)
                    elif (matrix_rm.loc[module_to,module_from]==0) and (matrix_rs.loc[module_to,module_from]==0):
                        matrix_row.append('0')
                        matrix_math_row.append(0)
                    else:
                        matrix_row.append("{:.2f}".format(matrix_rm.loc[module_to,module_from]) + ' ± ' + "{:.2f}".format(matrix_rs.loc[module_to,module_from]))
                        matrix_math_row.append(matrix_rm.loc[module_to,module_from])
                matrix_text.append(matrix_row)
                matrix_math.append(matrix_math_row)
            matrix_text_df = pd.DataFrame(matrix_text,index=matrix_rm.index,columns=matrix_rm.columns)
            matrix_text_df.to_csv(filename.replace('_rm_','_text_').replace('xlsx','csv'))
            matrix_math_np = np.asarray(matrix_math)
            matrix_math_inv = -np.linalg.inv(matrix_math_np)
            matrix_math_inv_df = pd.DataFrame(matrix_math_inv,index=matrix_rm.index,columns=matrix_rm.columns)
            matrix_math_inv_df.to_csv(filename.replace('_rm_','_inv_').replace('xlsx','csv'))
            # filling comparison matrix
            for module_from in matrix_rm.columns:
                if module_from != 'DPD':
                    for module_to in matrix_rm.index:
                        if module_from != module_to:
                            connections_rm.at[module_from + ' -> ' + module_to,cell_line + ' + ' + ligand] = matrix_rm.loc[module_to,module_from]
                            connections_rs.at[module_from + ' -> ' + module_to,cell_line + ' + ' + ligand] = matrix_rs.loc[module_to,module_from]
connections_rm
```

```python
connections_text = []

for connection in connections_rm.index:
    con_row = []
    for condition in connections_rm.columns:
        con_row.append("{:.2f}".format(connections_rm.loc[connection,condition]) + ' ± ' + "{:.2f}".format(connections_rs.loc[connection,condition]))
    connections_text.append(con_row)

connections_text_df = pd.DataFrame(connections_text,index=connections_rm.index,columns=connections_rm.columns)
connections_text_df.to_csv('connections_comparison.csv')
connections_text_df
```
